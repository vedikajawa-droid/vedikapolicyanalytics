```python
import pandas as pd

# Load the deidentified dataset
df = pd.read_csv('dataset_deidentified.csv')

print(f"Shape: {df.shape[0]:,} rows x {df.shape[1]} columns")
print(f"\nColumn names:\n{list(df.columns)}")
print(f"\nFirst few rows:")
df.head()
```

```python
# Replace special missing value codes with NaN (do not drop rows)
# Based on codebook inspection:
#   -99 = Not applicable / skip pattern (found in ENROLL_LEVEL_* columns)
#   -98 = System missing / refused / skipped (found across survey question columns)
#   Blanks are already NaN after pd.read_csv

MISSING_CODES = [-99, -98]

df_clean = df.replace(MISSING_CODES, pd.NA)

# Verify: confirm no -98 or -99 remain
remaining = {code: (df_clean == code).sum().sum() for code in MISSING_CODES}
print("Remaining special codes after replacement:", remaining)
print(f"\nNaN cells before: {df.isna().sum().sum():,}")
print(f"NaN cells after:  {df_clean.isna().sum().sum():,}")
print(f"Rows unchanged:   {len(df_clean):,} (no rows dropped)")
```

```python
# ── Step 1: create family_id and respondent_type ──────────────────────────────
# Only rows ending in _parent or _child are paired observations.
# The 2,472 rows ending in _adult are standalone GenZ panelists (Sample_Source=1)
# and are excluded from the paired analysis.

paired_mask = df_clean['Unique_ID'].str.endswith(('_parent', '_child'), na=False)
df_paired = df_clean[paired_mask].copy()

df_paired['respondent_type'] = df_paired['Unique_ID'].str.extract(r'_(parent|child)$')[0]
df_paired['family_id']       = df_paired['Unique_ID'].str.replace(r'_(parent|child)$', '', regex=True)

print(f"Paired rows: {len(df_paired):,}  (excluded {(~paired_mask).sum():,} _adult rows)")
print(f"\nrespondent_type counts:\n{df_paired['respondent_type'].value_counts()}")

# ── Step 2: split into parent_data and child_data ─────────────────────────────
parent_data = df_paired[df_paired['respondent_type'] == 'parent'].copy()
child_data  = df_paired[df_paired['respondent_type'] == 'child'].copy()

print(f"\nparent_data: {len(parent_data):,} rows")
print(f"child_data:  {len(child_data):,} rows")

# ── Step 3: check parent/child observations per family ────────────────────────
family_counts = df_paired.groupby('family_id')['respondent_type'].value_counts().unstack(fill_value=0)
print(f"\nFamily composition summary:")
print(family_counts.value_counts().reset_index().rename(columns={'count': 'n_families'}))

# ── Step 4: merge parent and child data on family_id ─────────────────────────
# Suffix _p = parent columns, _c = child columns
merged = pd.merge(
    child_data.drop(columns='respondent_type'),
    parent_data.drop(columns='respondent_type'),
    on='family_id',
    suffixes=('_c', '_p'),
    how='inner'
)

print(f"\nMerged dataset: {merged.shape[0]:,} rows x {merged.shape[1]} columns")
print(f"Families represented: {merged['family_id'].nunique():,}")
merged.head()
```

```python
# ── Confirm merged dataset structure ──────────────────────────────────────────
n_rows      = len(merged)
n_families  = merged['family_id'].nunique()
family_freq = merged['family_id'].value_counts()
multi_child = family_freq[family_freq > 1]

print(f"Rows in merged dataset:      {n_rows:,}")
print(f"Unique family_ids:           {n_families:,}")
print(f"1-to-1 match (rows==families): {n_rows == n_families}")
print()
if len(multi_child) > 0:
    print(f"Families with multiple children: {len(multi_child)}")
    print(multi_child.reset_index().rename(columns={'count': 'n_children', 'family_id': 'family_id'}))
else:
    print("No family_id appears more than once — every row is a unique child-parent pair.")
```
